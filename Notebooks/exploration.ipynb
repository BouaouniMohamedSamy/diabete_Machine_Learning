```python
# 1_exploration.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 📥 Charger les données
df = pd.read_csv('data/diabetes.csv')

# 👀 Aperçu du dataset
print("Dimensions :", df.shape)
display(df.head())

print("\n Valeurs manquantes :\n")
print(df.isnull().sum())

print("\n Statistiques :\n")
display(df.describe())

plt.figure(figsize=(6,4))
sns.countplot(x='Outcome', data=df, palette='coolwarm')
plt.title('Distribution des classes (0 = Non Diabétique, 1 = Diabétique)')
plt.show()

# 🔍 Corrélations entre les variables
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='RdBu_r', center=0)
plt.title('Matrice de corrélation')
plt.show()

df.hist(bins=20, figsize=(15, 10), color='steelblue')
plt.tight_layout()
plt.suptitle('Distribution des variables', fontsize=18, y=1.02)
plt.show()
```
